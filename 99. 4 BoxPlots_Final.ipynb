```python
import numpy as np

# Extracting the interquartile range (Q2) and (Q3 - Q1) for each metric and algorithm
medians = {
    'mae': [np.percentile(scores, 50)  for scores in [mae_test_knn, mae_test_svm, mae_test_rf, mae_test_et, mae_test_gbr, mae_test_xgboost, mae_test_catboost, mae_test_lgb, mae_test_ann]],
    'mape': [np.percentile(scores, 50) for scores in [mape_test_knn, mape_test_svm, mape_test_rf, mape_test_et, mape_test_gbr, mape_test_xgboost, mape_test_catboost, mape_test_lgb, mape_test_ann]],
    'rmse': [np.percentile(scores, 50) for scores in [rmse_test_knn, rmse_test_svm, rmse_test_rf, rmse_test_et, rmse_test_gbr, rmse_test_xgboost, rmse_test_catboost, rmse_test_lgb, rmse_test_ann]],
    'r2': [np.percentile(scores, 50)   for scores in [r2_test_knn, r2_test_svm, r2_test_rf, r2_test_et, r2_test_gbr, r2_test_xgboost, r2_test_catboost, r2_test_lgb, r2_test_ann]]
}

max_min_diff = {
    'mae': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [mae_test_knn, mae_test_svm, mae_test_rf, mae_test_et, mae_test_gbr, mae_test_xgboost, mae_test_catboost, mae_test_lgb, mae_test_ann]],
    'mape': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [mape_test_knn, mape_test_svm, mape_test_rf, mape_test_et, mape_test_gbr, mape_test_xgboost, mape_test_catboost, mape_test_lgb, mape_test_ann]],
    'rmse': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [rmse_test_knn, rmse_test_svm, rmse_test_rf, rmse_test_et, rmse_test_gbr, rmse_test_xgboost, rmse_test_catboost, rmse_test_lgb, rmse_test_ann]],
    'r2': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [r2_test_knn, r2_test_svm, r2_test_rf, r2_test_et, r2_test_gbr, r2_test_xgboost, r2_test_catboost, r2_test_lgb, r2_test_ann]]
}

# Rank algorithms
rankings = {}
for metric in ['mae', 'mape', 'rmse', 'r2']:
    if metric in ['r2']:
        rankings[metric + '_median'] = np.argsort(medians[metric])[::-1] + 1
    else:
        rankings[metric + '_median'] = np.argsort(medians[metric]) + 1

    rankings[metric + '_max_min_diff'] = np.argsort(max_min_diff[metric]) + 1

# Assign scores and calculate total scores
scores = {algorithm: {'mae': 0, 'mape': 0, 'rmse': 0, 'r2': 0, 'total': 0} for algorithm in algorithm_names}
for metric in ['mae', 'mape', 'rmse', 'r2']:
    for i, algorithm in enumerate(algorithm_names):
        scores[algorithm][metric + '_median'] = rankings[metric + '_median'][i]
        scores[algorithm][metric + '_max_min_diff'] = rankings[metric + '_max_min_diff'][i]
        scores[algorithm]['total'] += scores[algorithm][metric + '_median'] + scores[algorithm][metric + '_max_min_diff']

# Rank algorithms based on total score
total_rankings = np.argsort([scores[algorithm]['total'] for algorithm in algorithm_names])[::-1] + 1

# Display results
print("Ranking:")
for metric in ['mae', 'mape', 'rmse', 'r2']:
    print(f"\n{metric.capitalize()} Rankings:")
    for i, algorithm in enumerate(algorithm_names):
        print(f"{algorithm}: Median({rankings[metric + '_median'][i]}) Max_Min_Diff({rankings[metric + '_max_min_diff'][i]})")

print("\nTotal Rankings based on Total Score:")
for i, algorithm in enumerate(algorithm_names):
    print(f"{algorithm}: {total_rankings[i]} ({scores[algorithm]['total']})")
```

```python
import numpy as np
import pandas as pd

# Extracting Q2 and (Q3 - Q1) for each metric and algorithm
q2_values = {
    'mae': [np.percentile(scores, 50) for scores in [mae_test_knn, mae_test_svm, mae_test_rf, mae_test_et, mae_test_gbr, mae_test_xgboost, mae_test_catboost, mae_test_lgb, mae_test_ann]],
    'mape': [np.percentile(scores, 50) for scores in [mape_test_knn, mape_test_svm, mape_test_rf, mape_test_et, mape_test_gbr, mape_test_xgboost, mape_test_catboost, mape_test_lgb, mape_test_ann]],
    'rmse': [np.percentile(scores, 50) for scores in [rmse_test_knn, rmse_test_svm, rmse_test_rf, rmse_test_et, rmse_test_gbr, rmse_test_xgboost, rmse_test_catboost, rmse_test_lgb, rmse_test_ann]],
    'r2': [np.percentile(scores, 50) for scores in [r2_test_knn, r2_test_svm, r2_test_rf, r2_test_et, r2_test_gbr, r2_test_xgboost, r2_test_catboost, r2_test_lgb, r2_test_ann]]
}

q3_q1_diff = {
    'mae': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [mae_test_knn, mae_test_svm, mae_test_rf, mae_test_et, mae_test_gbr, mae_test_xgboost, mae_test_catboost, mae_test_lgb, mae_test_ann]],
    'mape': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [mape_test_knn, mape_test_svm, mape_test_rf, mape_test_et, mape_test_gbr, mape_test_xgboost, mape_test_catboost, mape_test_lgb, mape_test_ann]],
    'rmse': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [rmse_test_knn, rmse_test_svm, rmse_test_rf, rmse_test_et, rmse_test_gbr, rmse_test_xgboost, rmse_test_catboost, rmse_test_lgb, rmse_test_ann]],
    'r2': [np.percentile(scores, 75) - np.percentile(scores, 25) for scores in [r2_test_knn, r2_test_svm, r2_test_rf, r2_test_et, r2_test_gbr, r2_test_xgboost, r2_test_catboost, r2_test_lgb, r2_test_ann]]
}

# Define a function to calculate scores based on ranking
def calculate_score(metric_values, reverse=False):
    scores = pd.Series(metric_values).rank(ascending=not reverse)
    return pd.cut(scores, bins=9, labels=[i for i in range(1, 10)])

# Create new dictionaries to store scores
q2_scores = {}
q3_q1_diff_scores = {}

# Calculate scores for Q2
for metric in q2_values:
    reverse_score = metric == 'r2'
    q2_scores[metric + '_Score'] = calculate_score(q2_values[metric], reverse=reverse_score)

# Calculate scores for Q3-Q1 Difference
for metric in q3_q1_diff:
    q3_q1_diff_scores[metric + '_Score'] = calculate_score(q3_q1_diff[metric])

# Convert tuples to lists
q2_scores = {key: list(values) for key, values in q2_scores.items()}
q3_q1_diff_scores = {key: list(values) for key, values in q3_q1_diff_scores.items()}
# Convert dictionaries to DataFrames
q2_df = pd.DataFrame({**{'Algorithm': algorithm_names}, **q2_scores})
q3_q1_diff_df = pd.DataFrame({**{'Algorithm': algorithm_names}, **q3_q1_diff_scores})

def sum_scores(scores_dict):
    total_score = pd.DataFrame({
        'Algorithm': algorithm_names,
        'Total_Score': pd.concat([pd.Series(scores_dict[key], name=key) for key in scores_dict], axis=1).sum(axis=1)
    })
    return total_score.sort_values(by='Total_Score')

# Sum the scores for Q2 values
q2_sum_df = sum_scores(q2_scores)

# Sum the scores for Q3-Q1 Difference values
q3_q1_diff_sum_df = sum_scores(q3_q1_diff_scores)

# Merge Q2 and Q3-Q1 Difference total scores
total_scores_df = pd.merge(q2_sum_df, q3_q1_diff_sum_df, on='Algorithm', suffixes=('_Q2', '_Q3-Q1'))

# Sum the total scores to get the combined score
total_scores_df['Total_Score'] = total_scores_df['Total_Score_Q2'] + total_scores_df['Total_Score_Q3-Q1']

# Sort the DataFrame based on the combined score
total_scores_df = total_scores_df.sort_values(by='Total_Score')


print(q2_df)
print(q3_q1_diff_df)
# Display the combined total scores DataFrame
print("Combined Total Scores:")
print(total_scores_df)
```
